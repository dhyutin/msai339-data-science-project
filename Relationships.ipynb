```python
import pandas as pd

# Assuming 'processed' is your DataFrame
# processed = pd.DataFrame(...)

# Calculate Pearson's and Spearman's correlation coefficients
pearson_correlation = processed.corr(method='pearson')
spearman_correlation = processed.corr(method='spearman')

# Define thresholds for extreme correlations
threshold_positive = 0.7
threshold_negative = -0.7

# Function to find extreme correlations
def find_extreme_correlations(correlation_matrix, threshold_positive, threshold_negative):
    extreme_positive = []
    extreme_negative = []
    
    # Iterate through the correlation matrix
    for col in correlation_matrix.columns:
        for index in correlation_matrix.index:
            if index != col:  # Skip self-correlation
                if correlation_matrix.loc[index, col] >= threshold_positive:
                    extreme_positive.append((index, col, correlation_matrix.loc[index, col]))
                elif correlation_matrix.loc[index, col] <= threshold_negative:
                    extreme_negative.append((index, col, correlation_matrix.loc[index, col]))
                    
    return extreme_positive, extreme_negative

# Find extreme correlations for both Pearson's and Spearman's
pearson_extreme_positive, pearson_extreme_negative = find_extreme_correlations(pearson_correlation, threshold_positive, threshold_negative)
spearman_extreme_positive, spearman_extreme_negative = find_extreme_correlations(spearman_correlation, threshold_positive, threshold_negative)

# Print results
print("PEARSONS")
print("Extreme Positive Correlations (Pearson):")
for pair in pearson_extreme_positive:
    print(f"{pair[0]} and {pair[1]}: {pair[2]}")

print("\nExtreme Negative Correlations (Pearson):")
for pair in pearson_extreme_negative:
    print(f"{pair[0]} and {pair[1]}: {pair[2]}")

print("_____________________________________")

print("SPEARMAN")

print("\nExtreme Positive Correlations (Spearman):")
for pair in spearman_extreme_positive:
    print(f"{pair[0]} and {pair[1]}: {pair[2]}")

print("\nExtreme Negative Correlations (Spearman):")
for pair in spearman_extreme_negative:
    print(f"{pair[0]} and {pair[1]}: {pair[2]}")
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming 'processed' is your DataFrame
# One-hot encoding for categorical variables
one_hot_encoded_data = pd.get_dummies(processed, columns=['region', 'sex', 'race', 'marst', 'educ99'], drop_first=True)

# Print the head of the one-hot encoded DataFrame to check its structure
print("One-hot Encoded Data:")
print(one_hot_encoded_data.head())

# Ensure numeric columns are converted to binary if necessary
# For example, you can create binary indicators for other relevant numeric columns
if 'incwage' in one_hot_encoded_data.columns:
    one_hot_encoded_data['high_income'] = (one_hot_encoded_data['incwage'] > 50000).astype(int)
    one_hot_encoded_data.drop(columns=['incwage'], inplace=True)  # Drop the original column

# Ensure all columns are binary
one_hot_encoded_data = one_hot_encoded_data > 0  # Create a boolean DataFrame

# Print the head of the DataFrame to check binary conversion
print("Binary Encoded Data:")
print(one_hot_encoded_data.head())

# Apply the Apriori algorithm
frequent_itemsets = apriori(one_hot_encoded_data, min_support=0.05, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Filter rules for better readability
strong_rules = rules[(rules['confidence'] >= 0.7) & (rules['lift'] >= 1.0)]

# Print results
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nStrong Association Rules:")
print(strong_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Scatter plot for Support vs. Confidence
plt.figure(figsize=(10, 6))
plt.scatter(strong_rules['support'], strong_rules['confidence'], alpha=0.6)
plt.title('Support vs Confidence of Association Rules')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.axhline(y=0.7, color='r', linestyle='--', label='Confidence Threshold (0.7)')
plt.axvline(x=0.05, color='g', linestyle='--', label='Support Threshold (0.05)')
plt.legend()
plt.grid()
plt.show()

# Bar chart for top 10 rules by lift
top_rules = strong_rules.nlargest(10, 'lift')
plt.figure(figsize=(12, 6))
sns.barplot(x='lift', y=top_rules.index, data=top_rules, palette='viridis')
plt.title('Top 10 Association Rules by Lift')
plt.xlabel('Lift')
plt.ylabel('Rules')
plt.show()
```
